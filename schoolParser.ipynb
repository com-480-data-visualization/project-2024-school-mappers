#### A parser to structure the school dataframe to JSON for the REACT website


```python
import pandas as pd
import json

df = pd.read_csv('schools_3.csv')

structured_data = []
```

```python
grouped = df.groupby('schoolCountry').apply(lambda x: x.to_dict(orient='records')).to_dict()

# Convert the grouped data to JSON
json_data = json.dumps(grouped, indent=4)

# Save the JSON data to a file
with open('grouped_schools_3.json', 'w') as json_file:
    json_file.write(json_data)

print("JSON file has been created.")
```
